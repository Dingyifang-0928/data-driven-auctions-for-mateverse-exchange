```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
max_frequency = [15, 23, 15, 15, 6, 5, 3, 2, 0, 1]
# max_frequency = [i/100 for i in max_frequency]


range_boundary = np.linspace(0,0.5,11,retstep = False)
range_boundary = np.delete(range_boundary,0,axis = 0)
#
plt.rcParams['font.size'] = 16
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
plt.rcParams['figure.dpi'] = 300
plt.rcParams['figure.figsize'] = (8,6)
plt.rc('font',family='Times New Roman')
plt.rcParams['mathtext.fontset'] = 'custom'
plt.rcParams['mathtext.rm'] = 'Times New Roman'  # 
plt.rcParams['mathtext.it'] = 'Times New Roman:italic'  # 
x_axis = range_boundary
print(x_axis)
#
x = np.arange(len(x_axis))
width = 0.98

#
max_x = x
balance_x = x

#
plt.bar(max_x,max_frequency,width=width,color="firebrick",label=r"$\it{e_{opt}}$")
plt.xticks(ticks = [-0.5,0.5, 1.5, 2.5, 3.5, 4.5, 5.5, 6.5, 7.5, 8.5, 9.5],
           labels=[0,0.05, 0.1 , 0.15, 0.2 , 0.25, 0.3 , 0.35, 0.4 , 0.45, 0.5 ])
#
for i in range(len(x_axis)):
    plt.text(max_x[i],max_frequency[i], max_frequency[i],va="bottom",ha="center",fontsize=14)
plt.xlabel('Range of prediction errors')
plt.ylabel('Proportion of test data distribution(%)')
plt.legend(loc="upper right") 

savefile = './pic/'
figname = 'ps-max-t-ratio'
plt.savefig(savefile+figname)
```
